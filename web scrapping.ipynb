```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/80.0.3987.162 Safari/537.36"
}

webpage = "https://www.ambitionbox.com/list-of-companies?page=1"

response = requests.get(webpage, headers=headers)
```

```python
html = requests.get(webpage, headers=headers).text
soup = BeautifulSoup(html, "lxml")
```

```python
company=soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
name = []
rating = []
reviews = []
ctype = []
hq = []
how_old = []
no_of_employee = []

for i in company:

    # Company name
    name.append(i.find('h2').get_text(strip=True))

    # Rating
    r = i.find('p', class_='companyCardWrapper-companyRatingValue')
    rating.append(r.get_text(strip=True) if r else None)

    # Reviews
    rev = i.find('a', class_='companyCardWrapper-companyRatingCount')
    reviews.append(rev.get_text(strip=True) if rev else None)

    # Company details (type, HQ, age, employees)
    info = i.find_all('span', class_='companyCardWrapper-info')

    ctype.append(info[0].get_text(strip=True) if len(info) > 0 else None)
    hq.append(info[1].get_text(strip=True) if len(info) > 1 else None)
    how_old.append(info[2].get_text(strip=True) if len(info) > 2 else None)
    no_of_employee.append(info[3].get_text(strip=True) if len(info) > 3 else None)

```
