# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv("healthcare_dataset.csv")
df.head()
```

# Data Cleaning

```python
df.info()

df.isnull().sum()

df.drop_duplicates(inplace=True)
```

# Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution of Patients")
plt.show()
```

# Disease Distribution

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Medical Condition', data=df)
plt.title("Most Common Medical Conditions")
plt.show()
```

# Gender vs Disease

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Gender', hue='Medical Condition', data=df)
plt.title("Disease Distribution by Gender")
plt.show()
```

 # Billing Analysis

```python
billing = df.groupby('Medical Condition')['Billing Amount'].mean().sort_values()

billing.plot(kind='barh', figsize=(10,6))
plt.title("Average Billing per Disease")
plt.show()
```

# Admission Type

```python
sns.countplot(x='Admission Type', data=df)
plt.title("Admission Type Distribution")
plt.show()
```

# Correlation Analysis

```python
numeric_cols = df.select_dtypes(include=np.number)

sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
