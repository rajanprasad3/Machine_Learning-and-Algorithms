```python
import pandas as pd
```

```python
df = pd.read_csv("supernova_dataset.csv")
```

```python
df.isnull().mean()*100
```

```python
df = df.drop(columns=['kepid'])
```

```python
df = df.drop(columns=['koi_num_transits'])
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df = df[df['koi_prad'].notnull()]
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
error_cols = [col for col in df.columns if 'err' in col]
df = df.drop(columns=error_cols)
```

```python
df['koi_disposition'] = df['koi_disposition'].map({
    'CONFIRMED': 1,
    'FALSE POSITIVE': 0
})
```

```python
df = df[df['koi_disposition'].isin([0,1])]
```

```python
df.to_csv("cleaned_supernova_dataset.csv", index=False)
```
